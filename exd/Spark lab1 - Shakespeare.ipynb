```python
def make_plural(word):
    return word + 's'

print make_plural('cat')
```

```python
plural_RDD = words_RDD.map(make_plural)
plural_RDD.collect()
```

```python
word_pairs = plural_RDD.map(lambda x: (x, 1))
word_pairs.collect()
```

```python
word_grouped = word_pairs.groupByKey()
for k, v in word_grouped.collect():
    print k, list(v)
```

```python
word_counts = word_pairs.reduceByKey(lambda x, y : x + y)
word_counts.collect()
```

```python
from operator import add

totoal_count = word_counts.map(lambda x: x[1]).reduce(lambda x, y: x + y)
average = totoal_count / float(word_counts.count())
round(average, 2)
```

```python
def count_words(word_list_rdd):
    word_pairs = word_list_rdd.map(lambda x: (x, 1))
    word_counts = word_pairs.reduceByKey(lambda x, y : x + y)
    return word_counts

count_words(words_RDD).collect()
```

```python
import re

def remove_punctuation(text):
    return re.sub(r'[^a-zA-Z0-9_ ]', '', text.strip().lower())
    
remove_punctuation(" The Elephant's 4 cats. ")
```

```python
from operator import add
shakespeare_RDD = sc.textFile('pg100.txt', 8)
```

```python
def top_word_count(shakespeare_RDD, top=15):
    words = shakespeare_RDD.filter(lambda x: x != '').map(remove_punctuation).flatMap(lambda x: x.split())
    word_pairs = words.map(lambda x: (x, 1))
    word_counts = word_pairs.reduceByKey(lambda x, y : x + y)
    return word_counts.takeOrdered(top, key = lambda x: -x[1])
    
top_word_count(shakespeare_RDD)
```
